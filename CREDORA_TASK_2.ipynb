```python
# Basic libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
%matplotlib inline
sns.set(style='whitegrid')
```

```python
df=pd.read_csv('/content/gender_submission.csv')
df.head()
```

```python
# Load training data
train_df = pd.read_csv('/content/train (3).csv')  # adjust path if needed
test_df = pd.read_csv('/content/test.csv')

# View the first few rows
train_df.head()
```

```python
# @title Pclass

from matplotlib import pyplot as plt
train_df['Pclass'].plot(kind='hist', bins=20, title='Pclass')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt

train_df['Age'].plot(kind='hist', bins=20, title='Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

```python
import seaborn as sns

train_df['Sex'].value_counts().plot(kind='barh', color=sns.color_palette('Dark2'))
plt.title('Number of Passengers by Gender')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

```python
train_df.info()
train_df.describe(include='all')
```

```python
# Total missing values
train_df.isnull().sum()
```

```python
# Fill missing 'Age' with median
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)

# Fill missing 'Embarked' with mode
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' (too many missing values)
train_df.drop('Cabin', axis=1, inplace=True)
```

```python
# Convert 'Sex' to binary
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' to numeric
train_df['Embarked'] = train_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Drop unnecessary columns
train_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.histplot(train_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Step 8d: Fare Distribution by Survival
sns.boxplot(x='Survived', y='Fare', data=train_df)
plt.title('Fare by Survival')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation calculation
numeric_train_df = train_df.select_dtypes(include=np.number)
sns.heatmap(numeric_train_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
